#  Importing libraries

```python

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python
data = pd.read_csv("satisfaction.csv")
```

# Data Exploration

```python
data.head(5)
```

```python
data.isnull().sum()
```

# Data Preprocessing

```python
data = data.drop("Online support",axis=1)
```

```python
data['Arrival Delay in Minutes'].fillna(data['Departure Delay in Minutes'], inplace=True)
```

```python
data.rename(columns={'satisfaction_v2': 'Satisfaction'}, inplace=True)
```

# Exploratory Data Analysis (EDA) and Visualizations

```python
plt.figure(figsize=(8, 6))
data['Satisfaction'].value_counts().plot(kind='bar')
plt.title('Distribution of Satisfaction Levels')
plt.xlabel('Satisfaction Levels')
plt.ylabel('Count')
plt.show()
```

Distribution of Satisfaction Levels

```python
plt.figure(figsize=(8, 6))
data['Customer Type'].value_counts().plot(kind='bar')
plt.title('Count of Passengers by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.show()
```

Count of Passengers by Customer Type

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Distribution of Ages

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Flight Distance'], bins=10, edgecolor='black')
plt.title('Distribution of Flight Distances')
plt.xlabel('Flight Distance')
plt.ylabel('Frequency')
plt.show()
```

Distribution of Flight Distances

```python
type_of_travel_counts = data['Type of Travel'].value_counts()
class_counts = data['Class'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(type_of_travel_counts, labels=type_of_travel_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Types of Travel')
plt.show()

plt.figure(figsize=(8, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Classes')
plt.show()
```

Proportion of Types of Travel

```python
customer_type_counts = data['Customer Type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(customer_type_counts, labels=customer_type_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Customer Types')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data.boxplot(column='Seat comfort', by='Gender')
plt.title('Distribution of Seat Comfort by Gender')
plt.xlabel('Gender')
plt.ylabel('Seat Comfort')
plt.show()

```

Relationship between Age and Seat Comfort

```python
plt.figure(figsize=(8, 6))
data.boxplot(column='Seat comfort', by='Class')
plt.title('Distribution of Seat Comfort by Class')
plt.xlabel('Class')
plt.ylabel('Seat Comfort')
plt.show()
```

Relationship between Flight Distance and Arrival Delay

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x=data.index, y='Departure Delay in Minutes')
plt.title('Trend of Departure Delay Over Time')
plt.xlabel('Time')
plt.ylabel('Departure Delay (Minutes)')
plt.show()
```
